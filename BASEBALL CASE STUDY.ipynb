```python
from scipy.stats import zscore
z=zscore(df)
z
```

```python
dsn = df[(z < 3).all(axis=1)]
```

```python
dsn.shape
```

```python
from scipy.stats import probplot
plt.figure(figsize=(30,45))
for i in range(17):
    plt.subplot(6,3,i+1)
    probplot(dsn[col[i]],plot=plt,dist='norm')
    plt.title(col[i])
    plt.tight_layout()
```
